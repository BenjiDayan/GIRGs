```python
def foo(x, y):
    print(x)
    def inner():
        if y:
            nonlocal x
        x = 5
        print(x)
    inner()
    print(x)

foo(3, True)

print('------')

foo(3, False)
```

```python
def foo(x, y):
    print(x)
    def inner():
        if y:
            nonlocal x
            x = 5
        else:
            x = 5
        print(x)
    inner()
    print(x)

foo(3, True)

print('------')

foo(3, False)
```

```python
def foo(x, non_local):
    print(x)
    def inner():
        nonlocal x
        temp = x
        x = 5
        print(x)
        if not non_local:
            x = temp
    inner()
    print(x)

foo(3, True)

print('------')

foo(3, False)
```

```python
# import imp
# imp.reload(feature_extractor)
import feature_extractor
fe = feature_extractor.FeatureExtractor(graph_dicts)
output = {}


for i in range(len(graph_dicts)):
    if graph_dicts[i]['Group'] != 'socfb':
        continue
    print(graph_dicts[i])

    # with io.capture_output() as captured:
    out = fe._execute_one_graph(graph_dicts[i])
    for thing in out:
        for key, val in thing.items():
            if not key in output:
                output[key] = [val]
            else:
                output[key].append(val)
```
